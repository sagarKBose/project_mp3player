```python
from pygame import mixer
import os
```

```python
import tkinter.messagebox
from tkinter import filedialog
```

```python
mixer.init() #initializing mixer

#for tkinter 
root=Tk()
root.title("Mujjiic")
root.iconbitmap(r'headphones.ico')
#root.geometry('360x210')

l1=Label(root,text='Play Mujjic')
l1.pack(pady=10)


midFrame=Frame(root)
midFrame.pack(padx=40,pady=30)


bottomFrame=Frame(root)
bottomFrame.pack(padx=10,pady=10)


#for play button
play_pic=PhotoImage(file='icons8-play-48.png')
def play_fun():
    try:  #checks if 'paused' variable is initialized or not
        paused
    except NameError:  #if not initialized, go to this block
        try:
            mixer.music.load(filename)
            mixer.music.play()
            statusbar['text']='Playing :'+os.path.basename(filename)
        except:
            tkinter.messagebox.showerror('File not found','OOps! It seems the song file is not loaded. Load your song first and then enjoy!!')
    else:     #if 'pause' is initialized go to this block
        mixer.music.unpause()
        #statusbar['text']='Playing :'+os.path.basename(filename)
        statusbar['text']='Music Resumed'
  
play_bt=Button(midFrame,image=play_pic,highlightthickness = 0, bd = 0,command=play_fun)
play_bt.grid(row=2,column=2)

#for stop button
def stop_fun():
    mixer.music.stop()
    statusbar['text']='Music Stopped'
    
stop_pic=PhotoImage(file='stop.png')
stop_bt=Button(midFrame,image=stop_pic,highlightthickness = 0, bd = 0,command=stop_fun)
stop_bt.grid(row=2,column=3)

#for pause button
def pause_fun():
    global paused
    paused=True
    mixer.music.pause()
    statusbar['text']='Music Paused'
    
pause_pic=PhotoImage(file='icons8-pause-52.png')
pause_bt=Button(midFrame,image=pause_pic,highlightthickness = 0, bd = 0,command=pause_fun)
pause_bt.grid(row=2,column=1)


#for rewind button
def rewind_fun():
    play_fun()
    statusbar['text']='Music Rewined'
rewind_pic=PhotoImage(file='icons8-rewind-26.png')
rewind_bt=Button(midFrame,image=rewind_pic,highlightthickness = 0, bd = 0,command=rewind_fun)
rewind_bt.grid(row=2,column=4)


muted=False
#for mute and volume button
def mute_fun():
    global muted
    if muted:
        mixer.music.set_volume(0.7)
        volume_bt.config(image=volume_pic)
        scale.set(70)
        muted=False
    else:
        mixer.music.set_volume(0)
        volume_bt.config(image=mute_pic)
        scale.set(0)
        muted=True
    
mute_pic=PhotoImage(file='mute.png')
volume_pic=PhotoImage(file='volume.png')
volume_bt=Button(bottomFrame,image=volume_pic,highlightthickness = 0, bd = 0,command=mute_fun)
volume_bt.grid(row=0,column=1)


#for volume slider
def set_vol_fun(val):
    volume=int(val)/100
    mixer.music.set_volume(volume)
    
scale=Scale(bottomFrame,from_=0,to=100,orient=HORIZONTAL,command=set_vol_fun)
scale.set(70)
mixer.music.set_volume(0.7)
scale.grid(row=0,column=2)

#creating a menubar 
menubar=Menu(root)
root.config(menu=menubar)

#creating a submenu FILE
def open_file():
    global filename
    filename=filedialog.askopenfilename()

subMenu=Menu(menubar,tearoff=0)
menubar.add_cascade(label='File', menu=subMenu)
subMenu.add_command(label='Open',command=open_file)
subMenu.add_command(label='Exit',command=root.destroy)

#creating a submenu HELP
def about_fun():
    tkinter.messagebox.showinfo('About Mujjiic','This app is developed using Python Tkinter by Sagar')
    
subMenu=Menu(menubar,tearoff=0)
menubar.add_cascade(label='Help', menu=subMenu)
subMenu.add_command(label='About Us',command=about_fun)




#creating statusbar
statusbar=Label(root,text='Welcome to Mujjiic',relief=SUNKEN,anchor=W)
statusbar.pack(side=BOTTOM,fill=X)



root.mainloop()
```
